```python
import plotly.graph_objects as go
from datetime import date, timedelta
import pandas as pd
import pickle
import plotly.express as px

m = pickle.load(open('demandpredicition.pkl', 'rb'))
def typeofday(ds):
        date = pd.to_datetime(ds)
        if (date.weekday() == 5 or date.weekday() == 6):
            return 0
        elif(date.weekday() == 4):
            return 1
        else:
            return 2
sdate=date(2020,1,1)
edate=date(2020,1,7)
dates = pd.date_range(sdate,edate-timedelta(days=1),freq='h')
future = pd.DataFrame(dates)
future.columns = ['ds']
future['ds'] = pd.to_datetime(future['ds'])  
future['typeofday'] = future['ds'].apply(typeofday)
forecast = m.predict(future)
#forecast[forecast["yhat"]<0] = 0
# forecast['ds'] = pd.to_datetime(forecast['ds'])  


# fig = go.Figure(go.Scatter(
#     x = forecast['ds'],
#     y = forecast['yhat'],
#     mode='lines',
#     name = "Predicción"
# ))

df = px.data.stocks()
fig = px.line(forecast, x='ds', y="yhat")
fig.show()

fig.update_layout(title_text="Predicción semanal demanda <b>Neural Prophet</b> BiciMAD", yaxis_title="Demand")
fig.show()
```
